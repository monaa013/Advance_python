```python
import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
from mpl_toolkits.basemap import Basemap
%matplotlib widget
```

```python
fig = plt.figure(figsize=(5, 5))
m = Basemap(projection='lcc', resolution='c', width=8E6, height=8E6, lat_0=12.58, lon_0=77.30)
m.etopo(scale=0.5, alpha=0.5)

x, y = m(77.30, 12.58)  # longitude, latitude of Karnataka
plt.plot(x, y, 'ok', markersize=5)
plt.text(x, y, 'Karnataka', fontsize=12,color='blue')
plt.title('Karnataka in World Map')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', width=1.2E6, height=1.2E6,
            lat_0=12.9716, lon_0=77.5946)

# Draw coastlines, countries, and states
m.drawcoastlines()
m.drawcountries()
m.drawstates()

m.drawparallels(range(10, 20, 1), labels=[1,0,0,0])
m.drawmeridians(range(75, 80, 1), labels=[0,0,0,1])

x, y = m(77.5946, 12.9716)  
m.plot(x, y, 'bo', markersize=8) 
plt.text(x, y, 'Bangalore', fontsize=12, color='red')  
plt.title('Bangalore in South India')
plt.show()
```

```python
fig=plt.figure(figsize=(8,8))
m=Basemap()
m.drawcoastlines(linewidth=1.0, linestyle='dashed', color='red')
m.drawrivers(linewidth=0.5,linestyle='solid',color='skyblue')
plt.title("Coastlines", fontsize=20)
x, y = m(77.5946, 12.9716)
plt.plot(x,y,'ok',markersize=5,color='black')
plt.text(x,y,'Karnataka',fontsize=18,color='blue');
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', width=1.2E6, height=1.2E6,
            lat_0=12.9716, lon_0=77.5946)

# Draw coastlines, countries, and states
m.drawcoastlines(linewidth=1.0, linestyle='dashed', color='green')
m.drawrivers(linewidth=0.5,linestyle='solid',color='skyblue')
m.drawcountries(linewidth=1.0, linestyle='solid', color='k')
m.drawstates()

m.drawparallels(range(10, 20, 1), labels=[1,0,0,0])
m.drawmeridians(range(75, 80, 1), labels=[0,0,0,1])

x, y = m(75.993629, 12.528591)  
m.plot(x, y, 'bo', markersize=8) 
plt.text(x, y, 'Kaveri River', fontsize=12, color='red')  
plt.title('Kaveri River in South India')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', width=1.2E6, height=1.2E6,
            lat_0=12.9716, lon_0=77.5946)

m.drawcountries(linewidth=1.0, linestyle='solid', color='k')
m.drawcoastlines(linewidth=1.0, linestyle='dashed', color='red')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawmeridians(range(0,360,20),color='k',linewidth=1.0,dashes=[4,4],labels=[0,0,1,1])
m.drawparallels(range(-90,100,10),color='k',linewidth=1.0,dashes=[4,4],labels=[1,0,0,0])

x, y = m(77.5946, 12.9716)   
m.plot(x, y, 'bo', markersize=8) 
plt.text(x, y, 'Kaveri River', fontsize=12, color='red')  
plt.title('Bangalore in South India')
plt.show()
```
