```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('UScereal.csv')

# Bar Chart (2 variables) with custom color
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='calories', data=df, color='pink')  
plt.title('Bar Chart (2 variables)')
plt.show()
```

```python
# Bar Chart (3 variables) with custom color palette
plt.figure(figsize=(12, 8))
sns.barplot(x='Name', y='calories', hue='protein', data=df, palette='Set2')  
plt.title('Bar Chart (3 variables)')
plt.show()
```

```python
# Scatter Plot (2 variables) with custom color
plt.figure(figsize=(10, 6))
plt.scatter(df['calories'], df['protein'], color='green')  
plt.title('Scatter Plot (2 variables)')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
# Scatter Plot (3 variables) with custom color
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['calories'], df['protein'], df['fat'], c='Blue')  # Customize the color
ax.set_title('Scatter Plot (3 variables)')
ax.set_xlabel('Calories')
ax.set_ylabel('Protein')
ax.set_zlabel('Fat')
plt.show()
```

```python
# Histogram (2 variables) with custom colors
plt.figure(figsize=(10, 6))
plt.hist([df['calories'], df['protein']], bins=10, alpha=0.7, label=['Calories', 'Protein'], color=['orange', 'purple'])  
plt.title('Histogram (2 variables)')
plt.legend()
plt.show()
```

```python
# Histogram (3 variables) with custom colors
plt.figure(figsize=(12, 8))
plt.hist([df['calories'], df['protein'], df['fat']], bins=10, alpha=0.7, label=['Calories', 'Protein', 'Fat'], color=['black', 'red', 'violet'])  # Customize the colors
plt.title('Histogram (3 variables)')
plt.legend()
plt.show()
```

```python
# Pie Chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(df['calories'].value_counts(), labels=df['calories'].unique(), autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightgreen'])  # Customize the colors
plt.title('Pie Chart')
plt.show()
```

```python
# Pair Plot with custom colors
sns.set(style="whitegrid")
sns.pairplot(df, palette='indigo')  # Customize the color palette
plt.suptitle('Pair Plot')
plt.show()
```

```python
# Bubble Plot with custom colors
plt.scatter(df['calories'], df['protein'], s=df['fat']*10, alpha=0.5, color='salmon')  # Customize the color
plt.title('Bubble Plot')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
# Box Plot with custom colors
plt.figure(figsize=(10, 6))
sns.boxplot(x='calories', y='protein', data=df, color='purple')  # Customize the color
plt.title('Box Plot')
plt.show()
```

```python
# Violin Plot with custom colors
plt.figure(figsize=(12, 8))
sns.violinplot(x='calories', y='protein', data=df, palette='pastel')  # Customize the color palette
plt.title('Violin Plot')
plt.show()
```

```python
# Area Plot with custom color
plt.figure(figsize=(10, 6))
plt.fill_between(df['calories'], df['protein'], color='yellow', alpha=0.4)  # Customize the color
plt.title('Area Plot')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
# Line Plot with custom color
plt.figure(figsize=(10, 6))
plt.plot(df['calories'], df['protein'], marker='o', color='brown')  # Customize the color
plt.title('Line Plot')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
#plots examples
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ir = pd.read_csv('Iris.csv')
#print(ir.head(10))
#print(ir.describe())
#ir[3:2]
#ir.iloc[:,5].unique()
#print(ir.groupby('Species').size())
#print(ir.groupby(['Species']).min())
setosa=ir[ir['Species']=='Iris-setosa']
versicolor=ir[ir['Species']=='Iris-versicolor']
virginica=ir[ir['Species']=='Iris-virginica']
data=setosa["SepalWidthCm"]
data.plot(kind="hist",label="setosa",color='r')
plt.title('Setosa SepalWidthCm - Histogram')
plt.xlabel('Sepalwidthcm')
plt.show()
```

```python
setosa.plot(x="SepalLengthCm",y="SepalWidthCm",kind="scatter",label='setosa',color='blue')
plt.title('Scatter Plot - Setosa')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
setosa.plot(y="PetalLengthCm",kind="line",label='setosa',color='pink')
plt.title('Scatter Plot - Setosa')
plt.xlabel('PetalLengthCm')
plt.show()
```

```python
setosa.plot(y="PetalLengthCm",kind="line",label='setosa',color='black',linewidth='1.5',ls=':',legend='setosa')
plt.title('Scatter Plot - Setosa')
plt.xlabel('PetalLengthCm')
plt.show()
```

```python
setosa=ir[ir['Species']=='Iris-setosa']
versicolor=ir[ir['Species']=='Iris-versicolor']
virginica=ir[ir['Species']=='Iris-virginica']
setosa["SepalLengthCm"].plot(kind="line",color='black',linewidth='1.5',ls=':')
versicolor["SepalLengthCm"].plot(kind="line",color='blue',linewidth='1.5',ls=':')
virginica["SepalLengthCm"].plot(kind="line",color='green',linewidth='1.5',ls=':')
plt.title('Line Graph')
plt.xlabel('SepalWidthCm')
plt.legend(['setosa','versicolor','virginica'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = pd.read_csv('Iris.csv')

x = data['SepalLengthCm']
y = data['SepalWidthCm']
z = data['PetalLengthCm']

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c=x + y + z, cmap='viridis', marker='o') 
ax.set_title('3D Scatter Plot')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')

cbar = plt.colorbar(ax.scatter(x, y, z, c=x + y + z, cmap='viridis', marker='o'))
cbar.set_label('Combined Value')

# Show the plot
plt.show()
```
