```python
import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
from mpl_toolkits.basemap import Basemap
%matplotlib widget

iris=pd.read_csv('Iris.csv')
print(iris.head(10))
#print(iris.info())
#print(iris.shape)
#print(iris.describe())

```

```python
fig=plt.figure()

ax=plt.axes(projection = '3d')

x=iris['PetalWidthCm']
y=iris['PetalLengthCm']
z=iris['SepalLengthCm']

ax.scatter(x,y,z, c='red')
ax.set_title('3D')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
fig=plt.figure()

ax=plt.axes(projection = '3d')
ax.view_init(elev=30, azim=45, roll=18)

x=iris['PetalWidthCm']
y=iris['PetalLengthCm']
z=iris['SepalLengthCm']

ax.scatter(x,y,z, c='red')
ax.set_title('3D')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
fig=plt.figure()

bx=plt.axes(projection = '3d')

x=iris['PetalWidthCm'].sort_values()
y=iris['PetalLengthCm'].sort_values()
z=iris['SepalLengthCm'].sort_values()

bx.plot3D(x,y,z, c='red')
bx.set_title('3D')
bx.set_xlabel('PetalWidthCm')
bx.set_ylabel('PetalLengthCm')
bx.set_zlabel('SepalLengthCm')
plt.show()
```

```python
setosa=iris[iris['Species']=='Iris-setosa']
versicolor=iris[iris['Species']=='Iris-versicolor']
virginica=iris[iris['Species']=='Iris-virginica']
```

```python
#mpl.rcParams['font.size']=10
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.bar(setosa['Id'],setosa['PetalWidthCm'],zs=1,zdir='x',color='blue',alpha=0.8)
ax.bar(versicolor['Id'],versicolor['PetalWidthCm'],zs=2,zdir='x',color='red',alpha=0.8)
ax.bar(virginica['Id'],virginica['PetalWidthCm'],zs=3,zdir='x',color='green',alpha=0.8)
plt.show()
```
